```python
import pandas as pd
import numpy as np
import random
```

```python
data = {'國家':["Taiwan","American","China","Japan","South Korea","Thailand"] ,
        '人口':np.random.randint(10**6,10**9,size=6)}
population = pd.DataFrame(data)

print(population)
```

```python
population.iloc[population["人口"].idxmax(),:]
```
